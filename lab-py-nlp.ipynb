Load the dataset of positive and negative tweets.

Combine them into a single list called ``all_tweets`` and create a corresponding list of labels called `labels`.

```python
all_tweets = positive_tweets + negative_tweets

# Create corresponding labels
# Let's use 1 = positive, 0 = neutral, -1 = negative
labels = [1]*len(positive_tweets) + [-1]*len(negative_tweets)
```

```python
# Print a sample tweet
print("Sample Tweet:", all_tweets[0])
print("Label:", labels[0])
```

### **Exercise 9: Preprocess Tweets**

Apply the custom preprocessing pipeline to the entire dataset of tweets. Store the result in ``preprocessed_tweets``.

```python
# Custom preprocessing function
def preprocess_text(text, method="lemmatize"):
    stop_words = set(stopwords.words('english'))
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove stop words and punctuation
    filtered_tokens = [
        word for word in tokens
        if word.lower() not in stop_words and word not in string.punctuation
    ]
    
    # Apply stemming or lemmatization
    if method == "stem":
        processed_tokens = [stemmer.stem(word) for word in filtered_tokens]
    else:  # default lemmatization
        processed_tokens = [lemmatizer.lemmatize(word.lower()) for word in filtered_tokens]
    
    return processed_tokens


# Combine all tweets
all_tweets = positive_tweets + negative_tweets

# Apply preprocessing to entire dataset
preprocessed_tweets = [preprocess_text(tweet, method="lemmatize") for tweet in all_tweets]
```

```python
# Print a sample preprocessed tweet
print("Preprocessed Tweets Sample:", preprocessed_tweets[0])
```

### **Exercise 10: Feature Extraction on Tweets**

Extract features from the preprocessed tweets using **Bag of Words** and **TF-IDF**. Store the results in ``X_bow`` and ``X_tfidf``, respectively.

```python
# Join tokens back into strings
tweets_for_vectorizer = [" ".join(tokens) for tokens in preprocessed_tweets]

# --- Bag of Words ---
bow_vectorizer = CountVectorizer()
X_bow = bow_vectorizer.fit_transform(tweets_for_vectorizer)

print("BoW Vocabulary:", bow_vectorizer.get_feature_names_out())
print("BoW Feature Matrix:\n", X_bow.toarray())

# --- TF-IDF ---
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(tweets_for_vectorizer)

print("\nTF-IDF Vocabulary:", tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Feature Matrix:\n", X_tfidf.toarray())
```
